```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cd=pd.read_csv('company_sales_data.csv')
```

```python
cd.head()
```

#Read Total profit of all months and show it using a line plot

```python
month=cd['month_number'].tolist()
profit=cd['total_profit'].tolist()
plt.plot(month,profit)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.show()
```

#Get Total profit of all months and show line plot with the following Style properties

```python
profit=cd['total_profit'].tolist()
month=cd['month_number'].tolist()
plt.plot(month,profit,color='r',marker='o',markerfacecolor='k',linestyle='--',linewidth='3',label='Profit data of all months')
plt.xlabel('Month Number')
plt.ylabel('Sold units number')
plt.legend(loc='lower right')
```

#Read all product sales data and show it  using a multiline plot

```python
cd.head()
```

```python
month=cd['month_number'].tolist()
fc=cd['facecream'].tolist()
fw=cd['facewash'].tolist()
tp=cd['toothpaste'].tolist()
bs=cd['bathingsoap'].tolist()
sp=cd['shampoo'].tolist()
mz=cd['moisturizer'].tolist()
plt.plot(month,fc,marker='o',label='Face creame sales data')
plt.plot(month,fw,marker='o',label='Face wash sales data')
plt.plot(month,tp,marker='o',label='Toothpaste sales data')
plt.plot(month,bs,marker='o',label='Bathing soap sales data')
plt.plot(month,sp,marker='o',label='Shampoo sales data')
plt.plot(month,mz,marker='o',label='Moisturizer sales data')
plt.xlabel('Month Number')
plt.ylabel('sale units in number')
plt.legend(loc='upper left')
```

#Read toothpaste sales data of each month and show it using a scatter plot

```python
month=cd['month_number'].tolist()
tp=cd['toothpaste'].tolist()
plt.scatter(month,tp,label='Toothpaste sales data')
plt.xlabel('Month number')
plt.ylabel('Number of units sales')
plt.legend(loc='upper left')
plt.grid(True,linewidth=1,linestyle='--',)
```

#Read face cream and facewash product sales data and show it using the bar chart

```python
month=cd['month_number'].tolist()
fc=cd['facecream'].tolist()
fw=cd['facewash'].tolist()
plt.bar(month,fc,width=0.25,label='Facecream sales data',align='edge')
plt.bar(month,fw,width=-0.25,label='Facewash sales data',align='edge')
plt.xlabel('Month number')
plt.ylabel('Number of units sales')
plt.legend(loc='upper left')
```

#Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

```python
month=cd['month_number'].tolist()
bs=cd['bathingsoap'].tolist()
plt.bar(month,bs)
plt.xlabel('Month number')
plt.ylabel('Number of units sales')
```

#Read the total profit of each month and show it using the histogram to see most common profit ranges

```python
profit_range=[150000,175000,200000,225000,250000,300000,350000]
profit=cd['total_profit'].tolist()
plt.hist(profit,profit_range,label='profit data')
plt.xlabel('Profit')
plt.ylabel('profit range')
plt.legend(loc='upper left')
```
